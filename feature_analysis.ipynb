```python
import pandas as pd
import pickle
```

```python
data = pickle.load(open('features.pkl', 'rb'))
```

```python
data
```

```python
Protocol = ['Charge_C-rate',
'Time_after_charge',
'Time_after_discharge',
'Max_voltage',
'Cycles_since_soft_short']

QV_curve = ['ΔQ_3→13_var',
'ΔQ_3→13_max',
'ΔQ_3→13_min',
'ΔQ_3→13_mean',
'ΔQ_3→13_ske',
'ΔQ_3→13_kur',
'ΔQ_3→i_var',
'ΔQ_3→i_max',
'ΔQ_3→i_min',
'ΔQ_3→i_mean',
'ΔQ_3→i_ske',
'ΔQ_3→i_kur']

Relaxation = ['C_Rest_V_var',
'C_Rest_V_max',
'C_Rest_V_min',
'C_Rest_V_mean',
'C_Rest_V_ske',
'C_Rest_V_kur',
'D_Rest_V_var',
'D_Rest_V_max',
'D_Rest_V_min',
'D_Rest_V_mean',
'D_Rest_V_ske',
'D_Rest_V_kur',
'Chr_rest_V_drop_10min']

Cycle_characteristic = ['Q_max/min',
'Q_max-Q_min',
'Q_initial',
'Current_SOH',
'Q_slope',
'Norm_Charge_time',
'Ini_charge_time',
'Norm_CC_time',
'Norm_CV_time',
'CV_CC_ratio',
'Ini_CV_CC_ratio',
'CE']

all_features = Protocol + QV_curve + Relaxation + Cycle_characteristic
```

# Feature 분류
Feature들을 카테고리별로 분류합니다:
- **Protocol**: 실험 프로토콜 관련 (C-rate, Rest 시간, 전압 등)
- **QV_curve**: 차분 용량 곡선(dQ) 통계
- **Relaxation**: Rest 구간 전압 통계
- **Cycle_characteristic**: 사이클 특성 (용량, SOH, 충전 시간 등)

```python
for col in data.columns:
    if col not in all_features:
        print(col)
```

# DataFrame 컬럼 분류
DataFrame의 컬럼들을 용도에 따라 3가지로 분류합니다:
- **cell_info**: 셀 식별 정보 (cell_id, cell_type 등)
- **label**: 예측 대상 (RUL, hard_short_cycle 등)
- **feature**: 학습에 사용되는 특징들

```python
# DataFrame 컬럼들을 cell_info, label, feature로 분류

# 1. Cell 정보 컬럼 (식별자, 메타데이터)
cell_info_cols = [col for col in data.columns if col in ['cell_id', 'cell_type', 'cycle_number', 'cur_cyc_i']]

# 2. Label 컬럼 (예측 대상)
label_cols = [col for col in data.columns if any(keyword in col.lower() for keyword in ['label', 'rul', 'rsl', 'hard_short', 'target', 'class'])]

# 3. Feature 컬럼 (학습에 사용되는 특징들) - all_features에 정의된 것들
feature_cols = [col for col in data.columns if col in all_features]

# 분류 결과 출력
print(f"Cell Info 컬럼 ({len(cell_info_cols)}개):")
print(cell_info_cols)
print(f"\nLabel 컬럼 ({len(label_cols)}개):")
print(label_cols)
print(f"\nFeature 컬럼 ({len(feature_cols)}개):")
print(feature_cols[:10], "...")  # 처음 10개만 출력

# 분류되지 않은 컬럼 확인
all_classified = set(cell_info_cols + label_cols + feature_cols)
unclassified = [col for col in data.columns if col not in all_classified]
print(f"\n분류되지 않은 컬럼 ({len(unclassified)}개):")
print(unclassified)
```

```python
# DataFrame을 분류된 컬럼별로 분리

cell_info_df = data[cell_info_cols] if cell_info_cols else pd.DataFrame()
label_df = data[label_cols] if label_cols else pd.DataFrame()
feature_df = data[feature_cols] if feature_cols else pd.DataFrame()

print("=" * 60)
print("분리된 DataFrame 정보:")
print("=" * 60)
print(f"\n1. cell_info_df: {cell_info_df.shape}")
print(f"   컬럼: {list(cell_info_df.columns)}")

print(f"\n2. label_df: {label_df.shape}")
print(f"   컬럼: {list(label_df.columns)}")

print(f"\n3. feature_df: {feature_df.shape}")
print(f"   컬럼 수: {len(feature_df.columns)}")
print(f"   Feature 카테고리별 개수:")
print(f"   - Protocol: {len([f for f in feature_df.columns if f in Protocol])}")
print(f"   - QV_curve: {len([f for f in feature_df.columns if f in QV_curve])}")
print(f"   - Relaxation: {len([f for f in feature_df.columns if f in Relaxation])}")
print(f"   - Cycle_characteristic: {len([f for f in feature_df.columns if f in Cycle_characteristic])}")
```

```python
# 예시: 각 DataFrame 확인
print("=" * 60)
print("Cell Info DataFrame 샘플:")
print("=" * 60)
display(cell_info_df.head() if not cell_info_df.empty else "비어있음")

print("\n" + "=" * 60)
print("Label DataFrame 샘플:")
print("=" * 60)
display(label_df.head() if not label_df.empty else "비어있음")

print("\n" + "=" * 60)
print("Feature DataFrame 샘플 (처음 5개 컬럼):")
print("=" * 60)
display(feature_df.iloc[:, :5].head() if not feature_df.empty else "비어있음")
```

```python
# 2레벨 MultiIndex 생성 (cell_info, label, feature)
multi_columns = []

for col in data.columns:
    if col in cell_info_cols:
        multi_columns.append(('cell_info', col))
    elif col in label_cols:
        multi_columns.append(('label', col))
    elif col in feature_cols:
        multi_columns.append(('feature', col))
    else:
        multi_columns.append(('other', col))

# MultiIndex 적용
data.columns = pd.MultiIndex.from_tuples(multi_columns)

print("MultiIndex 적용 완료!")
print("\n컬럼 구조:")
print(data.columns)
print("\n상위 레벨 카테고리:")
print(data.columns.get_level_values(0).unique())
```
